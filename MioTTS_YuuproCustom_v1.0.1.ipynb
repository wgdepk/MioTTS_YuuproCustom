```python
import gradio as gr
import httpx
import json
import base64
import io
import numpy as np
import soundfile as sf
import subprocess as sp
import tempfile
import os
import sys
import shutil
from pathlib import Path
from datetime import datetime

API_BASE = f"http://localhost:{API_PORT}"
LOCAL_PRESETS = "/content/MioTTS-Inference/presets"
LOCAL_PRESETS_AUDIO = "/content/presets_audio"
DEFAULT_PRESETS = {"jp_female", "jp_male", "en_female", "en_male"}


# ==========================================
# 📖 辞書機能（Google Drive保存）
# ==========================================
def load_dictionary():
    if os.path.exists(DRIVE_DICT):
        with open(DRIVE_DICT, "r", encoding="utf-8") as f:
            entries = json.load(f)
        return "\n".join(f"{k}={v}" for k, v in entries.items())
    return ""

def save_dictionary(dict_text):
    entries = parse_dictionary(dict_text)
    with open(DRIVE_DICT, "w", encoding="utf-8") as f:
        json.dump(entries, f, ensure_ascii=False, indent=2)
    return f"✅ {len(entries)}件を Google Drive に保存しました"

def parse_dictionary(dict_text):
    entries = {}
    if not dict_text or not dict_text.strip():
        return entries
    for line in dict_text.strip().split("\n"):
        line = line.strip()
        if not line or "=" not in line:
            continue
        key, _, value = line.partition("=")
        key, value = key.strip(), value.strip()
        if key and value:
            entries[key] = value
    return entries

def apply_dictionary(text, dict_text):
    for key, value in parse_dictionary(dict_text).items():
        text = text.replace(key, value)
    return text


# ==========================================
# 🎚️ 話速コントロール（ffmpeg方式）
# ==========================================
def apply_speed(input_path, speed_rate):
    if abs(speed_rate - 1.0) < 0.01:
        return input_path
    output_path = input_path.replace(".wav", "_speed.wav")
    sp.run(
        ["ffmpeg", "-y", "-i", input_path,
         "-filter:a", f"atempo={speed_rate}", output_path],
        capture_output=True,
    )
    return output_path if os.path.exists(output_path) else input_path


# ==========================================
# 📥 プリセット管理
# ==========================================
def get_preset_list():
    presets_path = Path(LOCAL_PRESETS)
    if presets_path.exists():
        return sorted([p.stem for p in presets_path.glob("*.pt")])
    return []

def get_custom_preset_list():
    """カスタムプリセットのみ取得（デフォルト4種を除く）"""
    return [p for p in get_preset_list() if p not in DEFAULT_PRESETS]

def get_preset_audio_path(preset_name):
    """プリセットに対応するサンプル音声のパスを返す"""
    audio_path = os.path.join(LOCAL_PRESETS_AUDIO, f"{preset_name}.wav")
    if os.path.exists(audio_path):
        return audio_path
    return None

def on_preset_select(preset_name):
    """音声生成タブ: プリセット選択時にプレビュー音声を更新"""
    if not preset_name or preset_name in DEFAULT_PRESETS:
        return gr.update(value=None, visible=False)
    audio_path = get_preset_audio_path(preset_name)
    if audio_path:
        return gr.update(value=audio_path, visible=True)
    return gr.update(value=None, visible=False)

def register_preset(audio_path, preset_name):
    """新規プリセットを登録する"""
    if not audio_path:
        return "❌ 音声ファイルをアップロードしてください", gr.update(), gr.update()
    if not preset_name or not preset_name.strip():
        return "❌ プリセット名を入力してください", gr.update(), gr.update()

    safe_name = "".join(c for c in preset_name.strip() if c.isalnum() or c == "_")
    if not safe_name:
        return "❌ プリセット名は英数字とアンダースコアのみ", gr.update(), gr.update()

    processed_audio_path = audio_path
    temp_wav_file_name = None

    # Convert to WAV if not already WAV
    if Path(audio_path).suffix.lower() != '.wav':
        try:
            # Create a temporary WAV file
            temp_wav_file_name = tempfile.NamedTemporaryFile(suffix=".wav", delete=False).name

            # Use ffmpeg to convert to WAV
            conversion_result = sp.run(
                ["ffmpeg", "-y", "-i", audio_path, temp_wav_file_name],
                capture_output=True, text=True, timeout=60,
            )
            if conversion_result.returncode != 0:
                os.unlink(temp_wav_file_name) # Clean up temp file on error
                return (
                    f"❌ 音声ファイルの変換に失敗しました: {conversion_result.stderr}",
                    gr.update(), gr.update()
                )
            processed_audio_path = temp_wav_file_name
        except Exception as e:
            if temp_wav_file_name and os.path.exists(temp_wav_file_name):
                os.unlink(temp_wav_file_name)
            return f"❌ 音声ファイルの変換中にエラーが発生しました: {str(e)}", gr.update(), gr.update()

    # generate_preset.py でエンコード
    result = sp.run(
        [sys.executable, "scripts/generate_preset.py",
         "--audio", processed_audio_path, # Use the potentially converted WAV file
         "--preset-id", safe_name,
         "--output-dir", LOCAL_PRESETS],
        cwd="/content/MioTTS-Inference",
        capture_output=True, text=True, timeout=60,
    )

    preset_file = Path(LOCAL_PRESETS) / f"{safe_name}.pt"
    if not preset_file.exists():
        error_msg = result.stderr[-300:] if result.stderr else result.stdout[-300:]
        if temp_wav_file_name and os.path.exists(temp_wav_file_name):
            os.unlink(temp_wav_file_name) # Clean up temp file on error
        return f"❌ 登録失敗:\n{error_msg}", gr.update(), gr.update()

    # 元音声をローカルに保存 (processed_audio_path is now always WAV)
    local_audio = os.path.join(LOCAL_PRESETS_AUDIO, f"{safe_name}.wav")
    shutil.copy2(processed_audio_path, local_audio)

    # Google Driveにバックアップ（.pt + .wav）
    shutil.copy2(str(preset_file), DRIVE_PRESETS)
    shutil.copy2(local_audio, DRIVE_PRESETS_AUDIO)

    # Clean up the temporary WAV file if it was created
    if temp_wav_file_name and os.path.exists(temp_wav_file_name):
        os.unlink(temp_wav_file_name)

    new_all = get_preset_list()
    new_custom = get_custom_preset_list()
    return (
        f"✅ '{safe_name}' を登録しました（Drive にもバックアップ済み）",
        gr.update(choices=new_all, value=safe_name),
        gr.update(choices=new_custom),
    )

def on_manage_select(preset_name):
    """管理タブ: プリセット選択時に音声と名前をロード"""
    if not preset_name:
        return gr.update(value=None), gr.update(value="")
    audio_path = get_preset_audio_path(preset_name)
    return gr.update(value=audio_path), gr.update(value=preset_name)

def rename_preset(old_name, new_name):
    """プリセットの名前を変更する"""
    if not old_name:
        return "❌ プリセットを選択してください", gr.update(), gr.update(), gr.update(value=None), gr.update(value="")
    if not new_name or not new_name.strip():
        return "❌ 新しい名前を入力してください", gr.update(), gr.update(), gr.update(), gr.update()

    safe_new = "".join(c for c in new_name.strip() if c.isalnum() or c == "_")
    if not safe_new:
        return "❌ 名前は英数字とアンダースコアのみ", gr.update(), gr.update(), gr.update(), gr.update()
    if safe_new == old_name:
        return "ℹ️ 名前が同じです", gr.update(), gr.update(), gr.update(), gr.update()
    if Path(LOCAL_PRESETS, f"{safe_new}.pt").exists():
        return f"❌ '{safe_new}' は既に存在します", gr.update(), gr.update(), gr.update(), gr.update()

    # ローカル: .pt リネーム
    old_pt = Path(LOCAL_PRESETS) / f"{old_name}.pt"
    new_pt = Path(LOCAL_PRESETS) / f"{safe_new}.pt"
    if old_pt.exists():
        old_pt.rename(new_pt)

    # ローカル: 音声 リネーム
    old_audio = Path(LOCAL_PRESETS_AUDIO) / f"{old_name}.wav"
    new_audio = Path(LOCAL_PRESETS_AUDIO) / f"{safe_new}.wav"
    if old_audio.exists():
        old_audio.rename(new_audio)

    # Drive: 古いファイルを削除 → 新しい名前でコピー
    for drive_dir, ext in [(DRIVE_PRESETS, ".pt"), (DRIVE_PRESETS_AUDIO, ".wav")]:
        old_drive = Path(drive_dir) / f"{old_name}{ext}"
        if old_drive.exists():
            old_drive.unlink()
    if new_pt.exists():
        shutil.copy2(str(new_pt), DRIVE_PRESETS)
    if new_audio.exists():
        shutil.copy2(str(new_audio), DRIVE_PRESETS_AUDIO)

    new_all = get_preset_list()
    new_custom = get_custom_preset_list()
    audio_path = get_preset_audio_path(safe_new)
    return (
        f"✅ '{old_name}' → '{safe_new}' に名前を変更しました",
        gr.update(choices=new_all),
        gr.update(choices=new_custom, value=safe_new),
        gr.update(value=audio_path),
        gr.update(value=safe_new),
    )

def delete_preset(preset_name):
    """プリセットを削除する"""
    if not preset_name:
        return "❌ 削除するプリセットを選択してください", gr.update(), gr.update(), gr.update(value=None), gr.update(value="")
    if preset_name in DEFAULT_PRESETS:
        return "❌ デフォルトプリセットは削除できません", gr.update(), gr.update(), gr.update(), gr.update()

    # ローカル削除
    for path in [
        Path(LOCAL_PRESETS) / f"{preset_name}.pt",
        Path(LOCAL_PRESETS_AUDIO) / f"{preset_name}.wav",
    ]:
        if path.exists():
            path.unlink()

    # Drive削除
    for drive_dir, ext in [(DRIVE_PRESETS, ".pt"), (DRIVE_PRESETS_AUDIO, ".wav")]:
        drive_file = Path(drive_dir) / f"{preset_name}{ext}"
        if drive_file.exists():
            drive_file.unlink()

    new_all = get_preset_list()
    new_custom = get_custom_preset_list()
    new_value = new_custom[0] if new_custom else None
    return (
        f"✅ '{preset_name}' を削除しました（Driveからも削除済み）",
        gr.update(choices=new_all, value=new_all[0] if new_all else None),
        gr.update(choices=new_custom, value=new_value),
        gr.update(value=None),
        gr.update(value=""),
    )


# ==========================================
# 🎵 音声生成
# ==========================================
def generate_speech(
    text, preset, ref_audio, speed, dict_text, save_to_drive,
    temperature, top_p, max_tokens, rep_penalty
):
    if not text or not text.strip():
        raise gr.Error("テキストを入力してください")

    original_text = text.strip()
    processed_text = apply_dictionary(original_text, dict_text)

    processed_ref_audio = ref_audio
    temp_ref_wav_file = None

    if ref_audio is not None:
        # Convert ref_audio to WAV if not already WAV
        if Path(ref_audio).suffix.lower() != '.wav':
            try:
                temp_ref_wav_file = tempfile.NamedTemporaryFile(suffix=".wav", delete=False).name
                conversion_result = sp.run(
                    ["ffmpeg", "-y", "-i", ref_audio, temp_ref_wav_file],
                    capture_output=True, text=True, timeout=60,
                )
                if conversion_result.returncode != 0:
                    os.unlink(temp_ref_wav_file) # Clean up temp file on error
                    raise gr.Error(f"参照音声の変換に失敗しました: {conversion_result.stderr}")
                processed_ref_audio = temp_ref_wav_file
            except Exception as e:
                if temp_ref_wav_file and os.path.exists(temp_ref_wav_file):
                    os.unlink(temp_ref_wav_file)
                raise gr.Error(f"参照音声の変換中にエラーが発生しました: {str(e)}")

        with open(processed_ref_audio, "rb") as f:
            audio_bytes = f.read()
        ref_b64 = base64.b64encode(audio_bytes).decode("utf-8")
        reference = {"type": "base64", "data": ref_b64}
    elif preset:
        reference = {"type": "preset", "preset_id": preset}
    else:
        raise gr.Error("プリセットまたは参照音声を選択してください")

    payload = {
        "text": processed_text,
        "reference": reference,
        "llm": {
            "temperature": temperature,
            "top_p": top_p,
            "max_tokens": int(max_tokens),
            "repetition_penalty": rep_penalty,
        },
        "output": {"format": "base64"},
    }

    try:
        res = httpx.post(f"{API_BASE}/v1/tts", json=payload, timeout=120.0)
        res.raise_for_status()
    except httpx.HTTPStatusError as e:
        raise gr.Error(f"生成失敗: {e.response.status_code}")
    except Exception as e:
        raise gr.Error(f"APIエラー: {str(e)}")

    result = res.json()
    audio_b64 = result.get("audio", "")
    audio_bytes = base64.b64decode(audio_b64)
    audio_data, sr = sf.read(io.BytesIO(audio_bytes))

    tmp = tempfile.NamedTemporaryFile(suffix=".wav", delete=False)
    sf.write(tmp.name, audio_data, sr)
    final_path = apply_speed(tmp.name, speed)

    timings = result.get("timings", {})
    info = f"⏱️ 生成: {timings.get('total_sec', 0):.2f}秒"
    info += f" | 🔢 トークン: {result.get('token_count', '?')}"
    if abs(speed - 1.0) >= 0.01:
        info += f" | 🎚️ 話速: {speed:.1f}x"
    if original_text != processed_text:
        info += f"\n📖 辞書適用後: {processed_text}"

    if save_to_drive:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        short_text = original_text[:20].replace(" ", "_")
        drive_filename = "".join(c for c in f"{timestamp}_{short_text}.wav" if c.isalnum() or c in "_-")
        drive_path = os.path.join(DRIVE_OUTPUTS, drive_filename)
        shutil.copy2(final_path, drive_path)
        info += f"\n💾 Drive保存: {drive_filename}"

    # Clean up the temporary WAV file for reference audio if it was created
    if temp_ref_wav_file and os.path.exists(temp_ref_wav_file):
        os.unlink(temp_ref_wav_file)

    return final_path, info


# ==========================================
# 🌐 Gradio UIの構築
# ==========================================
initial_presets = get_preset_list()
initial_custom = get_custom_preset_list()
saved_dict = load_dictionary()

with gr.Blocks(
    title="MioTTS ゆうぷろカスタム_V1.0.0",
    theme=gr.themes.Soft(),
) as app:

    gr.Markdown("# 🎙️ MioTTS ゆうぷろカスタム_V1.0.0")
    gr.Markdown("🎚️ 話速 / 📖 辞書 / 📥 プリセット管理 / 💾 Drive連携")

    with gr.Tabs():
        # ===== タブ1: 音声生成 =====
        with gr.Tab("🎵 音声生成"):
            with gr.Row():
                with gr.Column(scale=3):
                    text_input = gr.Textbox(
                        label="📝 テキスト",
                        placeholder="音声にしたいテキストを入力...",
                        lines=4,
                    )
                    with gr.Row():
                        preset_dropdown = gr.Dropdown(
                            choices=initial_presets,
                            value=initial_presets[0] if initial_presets else None,
                            label="🔊 プリセット",
                        )
                        ref_audio_input = gr.Audio(
                            label="🎤 参照音声（任意）",
                            type="filepath",
                        )

                    # プリセットプレビュー
                    preset_preview = gr.Audio(
                        label="▶️ プリセット登録音声（プレビュー）",
                        type="filepath",
                        interactive=False,
                        visible=False,
                    )

                    speed_slider = gr.Slider(
                        minimum=0.5, maximum=2.0, value=1.0, step=0.1,
                        label="🎚️ 話速（0.5=ゆっくり / 1.0=標準 / 2.0=速い）",
                    )
                    save_drive_check = gr.Checkbox(
                        label="💾 生成音声を Google Drive に保存",
                        value=True,
                    )
                    generate_btn = gr.Button(
                        "🎵 音声を生成", variant="primary", size="lg",
                    )

                with gr.Column(scale=2):
                    audio_output = gr.Audio(label="🔈 生成音声", type="filepath")
                    info_output = gr.Textbox(label="ℹ️ 生成情報", interactive=False, lines=3)

            # プリセット選択 → プレビュー更新
            preset_dropdown.change(
                fn=on_preset_select,
                inputs=[preset_dropdown],
                outputs=[preset_preview],
            )

            with gr.Accordion("⚙️ 詳細設定（LLMパラメータ）", open=False):
                with gr.Row():
                    temperature = gr.Slider(0.0, 2.0, value=0.8, step=0.05, label="Temperature", info="低い=安定した生成 / 高い=多様だが不安定。途切れが気になる場合は0.6~0.7推奨")
                    top_p = gr.Slider(0.0, 1.0, value=1.0, step=0.05, label="Top-P", info="低い=高確率トークンのみ使用。0.9で不自然な間を減らせる場合あり")
                with gr.Row():
                    max_tokens = gr.Slider(100, 2000, value=700, step=50, label="Max Tokens", info="生成する音声の最大長（700=約28秒）。長い文章は増やしてください")
                    rep_penalty = gr.Slider(1.0, 1.5, value=1.0, step=0.05, label="Repetition Penalty", info="高い=同じパターンの繰り返しを抑制。途切れ防止には1.05~1.1推奨")
                with gr.Row():
                    reset_btn = gr.Button("🔄 初期設定に戻す", size="sm")

        # ===== タブ2: 読み辞書 =====
        with gr.Tab("📖 読み辞書"):
            gr.Markdown("### 📖 読み辞書")
            gr.Markdown("誤読する単語を登録 → 生成時に自動変換。`変換前=変換後` 形式で1行1エントリ")
            gr.Markdown("💾 辞書は Google Drive に保存されます")
            dict_input = gr.Textbox(
                label="辞書エントリ",
                placeholder="例:\nMioTTS=ミオティーティーエス\nColab=コラボ",
                lines=10,
                value=saved_dict,
            )
            with gr.Row():
                save_dict_btn = gr.Button("💾 辞書を保存", variant="primary")
                dict_status = gr.Textbox(label="状態", interactive=False, max_lines=1)
            save_dict_btn.click(
                fn=save_dictionary,
                inputs=[dict_input],
                outputs=[dict_status],
            )

        # ===== タブ3: プリセット管理 =====
        with gr.Tab("📥 プリセット管理"):

            # --- 新規登録 ---
            gr.Markdown("### 📥 新規プリセット登録")
            gr.Markdown("音声ファイルをプリセットとして登録（⚠️ **20秒以内**推奨）")
            with gr.Row():
                with gr.Column():
                    reg_audio = gr.Audio(label="🎤 登録する音声", type="filepath")
                    reg_name = gr.Textbox(
                        label="プリセット名（英数字・アンダースコア）",
                        placeholder="例: my_voice_01",
                    )
                    reg_btn = gr.Button("📥 プリセットとして登録", variant="primary")
                    reg_status = gr.Textbox(label="登録状態", interactive=False)

            gr.Markdown("---")

            # --- 登録済みプリセットの確認・編集 ---
            gr.Markdown("### ✏️ 登録済みプリセットの確認・編集")
            gr.Markdown("カスタムプリセットを選択して、音声の再生確認・名前変更・削除ができます。")

            manage_dropdown = gr.Dropdown(
                choices=initial_custom,
                label="📋 カスタムプリセット一覧",
                value=initial_custom[0] if initial_custom else None,
            )

            manage_audio = gr.Audio(
                label="▶️ 登録音声の再生",
                type="filepath",
                interactive=False,
            )

            with gr.Row():
                manage_new_name = gr.Textbox(
                    label="✏️ 新しいプリセット名",
                    placeholder="新しい名前を入力...",
                )
                rename_btn = gr.Button("✏️ 名前を変更", variant="primary")

            with gr.Row():
                del_btn = gr.Button("🗑️ このプリセットを削除", variant="stop")

            manage_status = gr.Textbox(label="操作結果", interactive=False)

            # 管理ドロップダウン選択 → 音声と名前をロード
            manage_dropdown.change(
                fn=on_manage_select,
                inputs=[manage_dropdown],
                outputs=[manage_audio, manage_new_name],
            )

            # 登録イベント
            reg_btn.click(
                fn=register_preset,
                inputs=[reg_audio, reg_name],
                outputs=[reg_status, preset_dropdown, manage_dropdown],
            )

            # 名前変更イベント
            rename_btn.click(
                fn=rename_preset,
                inputs=[manage_dropdown, manage_new_name],
                outputs=[manage_status, preset_dropdown, manage_dropdown, manage_audio, manage_new_name],
            )

            # 削除イベント
            del_btn.click(
                fn=delete_preset,
                inputs=[manage_dropdown],
                outputs=[manage_status, preset_dropdown, manage_dropdown, manage_audio, manage_new_name],
            )

    # リセットボタンのイベント
    def reset_params():
        return 0.8, 1.0, 700, 1.0

    reset_btn.click(
        fn=reset_params,
        outputs=[temperature, top_p, max_tokens, rep_penalty],
    )

    # 生成イベント
    generate_btn.click(
        fn=generate_speech,
        inputs=[
            text_input, preset_dropdown, ref_audio_input, speed_slider,
            dict_input, save_drive_check,
            temperature, top_p, max_tokens, rep_penalty,
        ],
        outputs=[audio_output, info_output],
    )

    gr.Markdown(
        "---\n"
        "### 📜 ライセンス\n"
        "| 対象 | ライセンス |\n"
        "|------|-----------|\n"
        "| 推論コード（MioTTS-Inference） | MIT License |\n"
        "| モデル（0.6B / 1.7B） | Apache 2.0 |\n"
        "| モデル（0.4B / 1.2B / 2.6B） | LFM Open License v1.0 |\n"
        "| モデル（0.1B） | Falcon-LLM License |\n"
        "| Ollama | MIT License |\n"
    )

print("="*70)
print("🎙️ MioTTS ゆうぷろカスタム_V1.0.0")
print("   🎚️ 話速 / 📖 辞書 / 📥 プリセット管理 / 💾 Drive連携")
print(f"   モデル: {GGUF_FILENAME}")
print(f"   Drive保存先: {DRIVE_BASE}")
print("="*70)
print("\n🌐 以下の公開URLをクリックしてWebUIにアクセスしてください\n")

app.launch(share=True, debug=True)
```
